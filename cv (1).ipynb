```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Q.1 Plot a graph of x^2, where x ranges from 0 to 9.

x = np.arange(0,10)
y = x*x
plt.plot(x,y,'r<-.',linewidth=0.5,markersize=8)
plt.xlabel('Number')
plt.ylabel('(Number)^2')
plt.title('Squares from 0 to 9')
```

```python
#Q.2 Line plot

x=np.arange(40,50)
y=np.arange(50,60)
plt.plot(x,y,'mo-',linewidth=0.5,markersize=8)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Q.3 Plot a graph showing sales trend of company 1 and 2

x = [1,2,3,4,5,6,7]                                                  # days of the week
Sales_1 = [160,150,140,145,175,165,180]                              # company 1 sales 
Sales_2 = [70,90,160,150,140,145,175]                                # company 2 sales 
plt.title("Sales trends over a week")
plt.plot(x, Sales_1,'ro-', label = 'Company 1 Sales')
plt.plot(x, Sales_2,'co-', label = 'Company 2 Sales')
plt.xlabel("Week days")
plt.ylabel("Company Sales")
plt.legend(loc='lower right')
plt.show()
```

```python
#Q.4 Creating 3 by 3 subplots
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,2,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplots(figsize=(14,14))

plt.subplot(3,3,1)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y1,'c_-',linewidth=0.5,markersize=8)

plt.subplot(3,3,2)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(y1,y2,'r|-',linewidth=0.5,markersize=8)

plt.subplot(3,3,3)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(y3,y5
         ,'gh-',linewidth=0.5,markersize=8)

plt.subplot(3,3,4)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y4,'bd-',linewidth=0.5,markersize=8)

plt.subplot(3,3,5)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y5,'yD-',linewidth=0.5,markersize=8)

plt.subplot(3,3,6)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(y4,y5,'m+-',linewidth=0.5,markersize=8)
```
